```python
webpage2=requests.get('https://www.themoviedb.org/search?query=best%20movies',headers=Headers)
```

```python
webpage2
```

```python
soup2=BeautifulSoup(webpage2.content,'html.parser')
soup2
```

```python
soup2.find("h2").get_text(strip=True)
```

```python
titles = soup2.find_all("h2")

for t in titles:
    print(t.get_text(strip=True))
```

```python
soup3=BeautifulSoup(webpage2.content,'html.parser')
soup3
date_tag = soup3.find("span", class_="release_date").get_text(strip=True)
date_tag

date_tag = soup3.find_all("span", class_="release_date")

for d in date_tag:
    print(d.get_text(strip=True))

```

```python
soup4=BeautifulSoup(webpage2.content,'html.parser')
soup4
description=soup4.find('div',class_='overview').get_text(strip=True)
description
description=soup4.find_all('div',class_='overview')
for d in description:
    print(d.get_text(strip=True))
```

```python
min_len = min(len(titles), len(date_tag), len(description))

titles = titles[:min_len]
date_tag = date_tag[:min_len]
description = description[:min_len]
```

```python
import pandas as pd


df = pd.DataFrame({
    "Title": titles,
    "Date": date_tag,
    "Description": description
})


path = r"C:\Users\dell\Documents\WebScraping\movies_data.xlsx"
df.to_excel(path, index=False)

print("✅ File saved successfully at:", path)
```
